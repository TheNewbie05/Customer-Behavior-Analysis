```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median))
```

```python
df['Review Rating'] = (
    df.groupby('Category')['Review Rating']
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns ={'purchase_amount_(usd)':'purchase_amount'})
```

Createing column age_group and dividing labels

```python
labels= ['Young adult','Adult','Middle aged','Senior']
df['age_group']= pd.qcut(df['age'], q=4,labels= labels)
```

```python
df['frequency_of_purchases'].head(50)
```


creating column of frequency_of_purchases_days


```python
frequency_mapping = { 
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' :90,
    'Annually' : 365,
    'Bi-Weekly' :14,
    'Every 3 Months' : 90
}

df['frequency_of_purchases_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
